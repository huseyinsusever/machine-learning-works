```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Ice_Cream_Sales_temperatures.csv")
df.head()
```

```python
# We fit and train the model
X = df[['Temperature']] 
y = df['Ice_Cream_Profits']
model = LinearRegression()
model.fit(X, y)

print("Model Coefficient (Slope):", model.coef_[0])
print("cut-off point (Intercept):", model.intercept_)



```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.7)
plt.xlabel("Temperature")
plt.ylabel("Price")
plt.plot(X, y, 'r--', lw=2)
plt.show()
```

```python
# Let's make an example prediction
new_termometer = 30 
predict_sales = model.predict([[new_termometer]])
print(f"\n{new_termometer} C degree sales degree Ice cream: {predict_sales[0]:.2f} TL")
```
